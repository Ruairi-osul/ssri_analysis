```python
import pandas as pd
from pathlib import Path
from plotnine import *
```

```python
data_dir = Path(".").absolute().parent / "data"

df = pd.read_csv(data_dir / "baseline.csv")
```

### Naive

```python
dfn = (
    df[df["group_name"].isin(
        ["acute_citalopram", 
         "acute_saline", 
         "shock","sham", 
         "acute_cit", 
         "acute_sal"]
    )]
)
```

```python
ggplot(dfn, aes(x=1, fill="cluster")) + geom_bar(position="stack") + theme_classic()
```
